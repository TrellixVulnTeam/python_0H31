```python
import re
def find_regexp(regexp, string):
    res = re.search(regexp, string)
    if str(res) != 'None':
        result = 1
    else:
        result = 0
    return result
```

```python
names = dt['Название']
patterns = [r'^сист.\D*бл',r'^сист \D*бл',r'^системный\D*бл',r'^ссистемный\D*бл',r'систтемный\D*бл',
            r'систменый\D*бл', r'^acer', r'desten', r'более',
            r'^моноблок', r'ноутбук', r'notebook', r'^pc', r'монитор', r'lcd монитор', r'офис\D*\+\D*монитор',
            r'рабочая станция hp dc7800cmt, монитор 22 hp tft',
            r'^пк',r'^pc',  r'^[к|k]омпьютер[\d|\+|\s]', r'неттоп', r'^morex']
pc_names_indexes = []

for i in range(len(names)):
    try:
        name = names[i].lower()
        name = name.lstrip()
        for pattern in patterns:
            if find_regexp(pattern, name) == 1:
                pc_names_indexes.append(i)
                break
    except KeyError:
        pass

pc_names_indexes
```
