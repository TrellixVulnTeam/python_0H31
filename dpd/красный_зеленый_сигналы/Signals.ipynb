```python
import pandas as pd
import numpy as np
```

```python
def import_excel(path,sheet):
    data = pd.ExcelFile(path).parse(sheet, header=0)
    return data
```

```python
dt = import_excel('full.xlsx','exportSD')
dt
```

```python
agents = list(dt['Контрагент (сотрудник)'].unique())
agents
```

```python
def get_descr(agent):
    return list(dt['Описание запроса TXT'][dt['Контрагент (сотрудник)']==agent])
```

```python
def get_dt(agent):
    dt_new = pd.DataFrame()
    dt_new = dt[dt['Контрагент (сотрудник)']==agent]
    cols = dt_new.columns.to_list()
    cols.remove('Контрагент (сотрудник)')
    cols.remove('Описание запроса TXT')
    dt_new.drop(columns=cols, inplace=True)
    return dt_new
```

```python
dt1 = get_dt('СПСР ParcelTrack')
dt1
```

```python
arr = list(dt1['Описание запроса TXT'])
# print(len(arr))
address=[]; state=[]; output=[];
for i in dt1['Описание запроса TXT']:
    st = i.split('\n')
    for k in st:
        if k.startswith('Address')==True:
            address.append(k[k.find(':')+1:].lstrip().rstrip())
        if k.startswith('State')==True:
            state.append(k[k.find(':')+1:].lstrip().rstrip())      
dt1['address'] = address
dt1['state'] = state
```
