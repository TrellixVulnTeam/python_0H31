```python
import re
def find_regexp(regexp, string):
    res = re.search(regexp, string)
    if str(res) != 'None':
        result = 1
    else:
        result = 0
    return result
```

```python
names = dt['Название']
patterns = [r'^сист\.\D*бл[о|\.]',r'^сист \D*бл',r'^системный\D*бл',r'^ссистемный\D*бл',r'систтемный\D*бл',
            r'систменый\D*бл', r'^acer', r'desten', r'более',
            r'^моноблок', r'ноутбук', r'notebook', r'^pc', r'^монитор', r'lcd монитор',r'^жк-монитор', r'офис\D*\+\D*монитор',
            r'рабочая станция hp dc7800cmt, монитор 22 hp tft',
            r'^пк',r'^pc',  r'^[к|k]омпьютер[\d|\+|\s]', r'неттоп', r'^morex']
pc_names_indexes = []

for i in range(len(names)):
    try:
        name = names[i].lower()
        name = name.lstrip()
        for pattern in patterns:
            if find_regexp(pattern, name) == 1:
                pc_names_indexes.append(i)
                break
    except KeyError:
        pass

pc_names_indexes
```

```python
# проверка каждого патерна
patterns = [r'^[к|k]омпьютер[\d|\+|\s]']
pc_names = []
for name in names:
    name = name.lower()
    name = name.lstrip()
    for pattern in patterns:
        if (find_regexp(pattern, name) == 1) and (name not in pc_names):
            pc_names.append(name)
pc_names
```

```python
import datetime
d = datetime.date(2021,1,1)
# k=0
for i in list(dt_pc['Дата ввода в эксплуата-цию'].index):
    try:
        parse_date = datetime.date(dt_pc['Дата ввода в эксплуата-цию'][i].year, dt_pc['Дата ввода в эксплуата-цию'][i].month, dt_pc['Дата ввода в эксплуата-цию'][i].day)
        diff = d-parse_date
        n_years = int(diff.days/365.18)
        dt_pc['срок использования'][i] = n_years
    except TypeError:
        pass
dt_pc
```

```python
def _zfill_5(column):
    dt_pc[column] = dt_pc[column].astype(str)
    for i in list(dt_pc[column].index):
        if dt_pc[column][i] != 'nan':
            string = dt_pc[column][i]
            dt_pc[column][i] = string[:len(string)-1].replace('.','').zfill(5)
        else:
            dt_pc[column][i] = dt_pc[column][i].replace('nan', '')
```

```python
_zfill_5('ответственный табельный номер')
dt_pc
```
