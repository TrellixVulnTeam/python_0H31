```python
import pandas as pd
import numpy as np
```

```python
def import_excel(path,sheet):
    data = pd.ExcelFile(path).parse(sheet, header=0)
    return data
```

```python
dt1 = import_excel('ask1.xlsx','exportSD')
dt2 = import_excel('ask2.xlsx','exportSD')
```

```python
dt1
```

```python
dt_all = pd.concat([dt1,dt2], ignore_index=True)
dt_all
```

```python
l =["установить контур", 'ошибка при входе в экстерн', 'просьба установить диадок', 'контур экстерн', 'доступ в диадок', 'доступ в экстерн', 'контур диадок', 'контур эдо', 'установить эцп', 'цифровую подпись', 'эцп воинова', 'сертификат', 'криптопро']
```

```python
dt = dt_all.iloc[:,[0,2,3]]
dt
```

```python
dt['search'] = np.nan
```

```python
for i in range(len(dt['Тема'])):
    search = []
    if dt['Тема'][i] ==dt['Тема'][i]:
        tema = dt['Тема'][i].lower()
        for phrase in l:
            if tema.find(phrase) != -1:
                print(phrase)
                search.append(phrase)
    if dt['Описание запроса TXT'][i] ==dt['Описание запроса TXT'][i]:
        des = dt['Описание запроса TXT'][i].lower()
        for phrase in l:
            if des.find(phrase) != -1:
                print(phrase)
                search.append(phrase)
    dt['search'][i] = ', '.join(search)
dt
```
