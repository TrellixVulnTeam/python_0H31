## Задача наименьшего разреза графа

__Сложность задач разбиения графа__

<em class="qs"></em> Выделение соообществ (community detection) в сети 

Свойства:
* <em class="cb">Разделенность.</em>  Высокая плотность связей внутри группы по сравнению со связями с внешними членами.
* ... (существет много альтернативных определений)

<em class="df"></em> __Разбиение графа__ (graph partition) - разбиение вершин графа на множество подмножеств (подграфов) вершин. 
* Обычно требуется, чтобы каждая вершина принадлежала хотя <em class="cb">одному и только одному подмножеству</em>. 
* Обычно для разбиения графа предяявляются <em class="cb">дополнительные требования</em>. 
* <em class="cb">Выделение сообществ</em> - частный случай решения задачи разбиение графа.

Комбинаторный взгляд на задачу разбиения графа:
* Количество вариантов разбиения графа из $N$ узлов на 2 подграфа: 
$$C_N^{N_1}=C_N^{N_2}=\frac{N!}{N_1!N_2!},\: N=N_1+N_2$$ 
* Количество всех возможных разбиений графа из $N$ узлов (количество разбиений множества на подмножества - число Белла): $B_n=\sum_{k=1}^n S(n,k)$, где $S(n,k)$ - количество неупорядоченных разбиений $n$-элементного множества на $k$ непустых подмножеств. $B_n$ можно рассчитать как:
$$B_{n+1}=C_n^k B_n, B_0=1$$ 
* $B_0=1,B_1=1,B_2=2,B_3=5,B_4=15,B_5=52,B_6=203,B_7=877,B_8=4\,140,B_9=21\,147,B_{10}=115\,975,\ldots,B_{15}=1\,382\,958\,545,\ldots,B_{20}=51\,724\,158\,235\,372,\ldots,B_{25}=4\,638\,590\,332\,229\,999\,353,\ldots$
* Как правило задачи разбияния графа являются NP-сложными  

__Разрез графа__


<em class="df"></em> __Разрез графа__ (cut) — множество рёбер, удаление которых делит граф на два изолированных подграфа.
* Полученный изолированный подграф, в частности, может быть отдельным узлом.

<em class="df"></em> __Наименьший разрез графа__ (minimum cut) - это минимальный в некотором смысле разрез.

Пусть имеется сеть из множества связей $L$ и узлов $N$, разбитых на 2 не пересекающихся подмножества: $N_1 \cup N_2 = N$. Рассмотрим задачи наименьшего разреза графа, поставленные с помощью различных минимизируемых функций $Q$:
* Graph cut:
$$Q=cut(N_1,N_2)=\sum_{i \in N_1, j \in N_2} l_{ij}$$
    * приводит разбиениям на несбалансированные подмножества  $N_1$ и $N_2$

__Пример сбалансированных и не сбалансированных разбиений__

<img src="clust2_02.png" alt="Пример разбиений" style="width: 400px;"/>
</center>

* Ratio cut:
$$Q=\frac{cut(N_1,N_2)}{\Vert N_1 \Vert}+\frac{cut(N_1,N_2)}{\Vert N_2 \Vert}$$
    * учитывает количество узлов в подмножествах $N_1$ и $N_2$    
* Обозначим как $Vol(N_1) = \sum_{i \in N_1, j \in N} l_{ij}=\sum_{i \in N_1} k_{i}$ сумму степеней узлов подмножества
* Normalized cut:
$$Q=\frac{cut(N_1,N_2)}{Vol(N_1)}+\frac{cut(N_1,N_2)}{Vol(N_2)}$$
    * учитывает количество связей в подмножествах $N_1$ и $N_2$    
* Quotient cut (conductance):
$$Q=\frac{cut(N_1,N_2)}{\min(Vol(N_1),Vol(N_2))}$$
    * учитывает количество связей в подмножествах $N_1$ и $N_2$    
    
Задачу наименьшего разреза графа можно рассматривать как <em class="cb">одну из постановок задачи выделения соообществ в сети</em> (количесво связей между сообествами должно быть мало!)

__Матрица Лапласа__

* Пусть имеется сеть из множества связей $L$ и узлов $N$, разбитых на 2 не пересекающихся подмножества: $N_1 \cup N_2 = N$.
* определим вектор $\mathbf{s}$ из $n$ компонент (индикаторный вектор):
$$s(i) = \begin{cases} \ \ +1 \text{, если } n_i \in N_1  \\ \ \ -1 \text{, если } n_i \in N_2 \end{cases}$$
* тогда:
$$\frac{1}{2}(1-s_i s_j)= \begin{cases} \ \ 0 \text{, если } n_i \text{ и } n_j \text{из одной группы}  \\ \ \ 1\text{, если } n_i \text{ и } n_j \text{из разных групп} \end{cases}$$
* тогда разрез $cut(N_1,N_2)$ можно записать следующим образом:
$$cut(N_1,N_2)=\frac{1}{4}\sum_{i,j}(1-s_i s_j)A_{ij}$$
* заметим, что:
$$\sum_{i,j}A_{ij}=\sum_{i}k_{i}=\sum_{i}s_i^2k_{i}=\sum_{i,j}s_i s_j k_{i} \delta_{ij}$$
* тогда: 
$$cut(N_1,N_2)=\frac{1}{4}\sum_{i,j}(1-s_i s_j)A_{ij}=\frac{1}{4}\sum_{i,j}s_i s_j(k_{i} \delta_{ij}-A_{ij})=\frac{1}{4} \mathbf{s^T}\mathbf{L}\mathbf{s}$$
где $\mathbf{L}$ — матрица Лапласа с элементами:
$$L_{ij} = \begin{cases} \ \ k_i \text{, если } i=j  \\ \ \ -1 \text{, если } i \ne j \text{ и между } i \text{ и } j \text{ есть ребро}\\ \ \ 0 \text{, иначе }\end{cases}$$
или $\mathbf{L}=\mathbf{D}-\mathbf{A}$ где $D_{ij}=diag(k_i)$




__Свойства матрицы Лапласа__

Матрица Лапласа:
$$L_{ij} = \begin{cases} \ \ k_i \text{, если } i=j  \\ \ \ -1 \text{, если } i \ne j \text{ и между } i \text{ и } j \text{ есть ребро}\\ \ \ 0 \text{, иначе }\end{cases}$$
или $\mathbf{L}=\mathbf{D}-\mathbf{A}$ где $D_{ij}=diag(k_i)$

__Пример матрицы Лапласа для графа "путь"__

* матрица симметричная <em class="hn"></em> из собственных векторов можно составить ортонормированный базис
* все собственные значения матрицы неотрицательны
* cумма по любой строке или по любому столбцу равна 0
* <em class="hn"></em> всегда существует нулевое собственное значение и отвечающий ему ссобственный вектор $(1,1, \ldots,1)$ (или нормированный собственный вектор:  $(1,1, \ldots,1)/\sqrt{n}$)

<em class="cb">__Лемма__</em>: для неориентированной сети с неотрицательными весами кратность $k$ нулевого собственного значения матрицы лапласа $L$ равна количеству связных компонент сети $N_1, \ldots, N_k$. Собственные пространство с.з. 0, над собственными векторами, отвечающимим с.з. 0 является оболочкой над индикаторными векторами этих компонент.

* Без ограничения общности узлы относящиеся к связным компонентам можно сгруппировать, тогда получится
* <em class="hn"></em> матрица инцидентности $\mathbf{A}$ будет иметь блочно-диагональный вид
* <em class="hn"></em> матрица Лапласа $\mathbf{L}$ будет иметь блочно-диагональный вид:
$$L = \begin{pmatrix}
L_1& & &\\
&L_2 & &\\
& & \ddots &\\
& & &L_3\\
\end{pmatrix}$$
* <em class="hn"></em> собственные вектора, отвечающие с.з. 0 и представляющие различные компоненты будут иметь 1 для узлов относящихся к компоненте, и 0 для остальных узлов.

__Матричная формулировка задачи наименьшего разреза__

Задачу наименьшего разреза графа можно сформулировать с использованием матрицы Лапаласа:
$$\underset{\mathbf{s}}{\min}Q(\mathbf{s})=\frac{\mathbf{s^T}\mathbf{L}\mathbf{s}}{4}$$
* $s_i=\pm 1$
* добавим условие сбалансированности классов: $\sum_{i}s_i=0$
* данная постановка является задачей целочисленной оптимизации и <em class="cb">является NP-сложной</em>

## Спектральный метод наименьшего разреза графа

__Непрерывный вариант задачи наименьшего разреза__

Т.к. найти точное решения для крупных сетей невозможно, применим один из подходов для снижения сложности задачи:
* <em class="cb">Приближенная оптимизация</em> - решение аналогичной задачи оптимизации с ослабленными условиями. Прежде всего рассматривается ослабление требования на получение целочисленного решения.

* Вместо вектора $\mathbf{s}$ с целочисленными значениямм: $s_i=\pm 1$ рассмотрим вектор $\mathbf{x}$ из $n$ действительных компонент.


* Задача непрерывной оптимизации имеет вид:
$$\underset{\mathbf{x}}{\min}Q(\mathbf{x})=\frac{\mathbf{x^T}\mathbf{L}\mathbf{x}}{4}$$
* при ограниченях:
    * $\sum_{i}x_i^2=n$
    * $\sum_{i}x_i=0$
    

* После нахождения вектора $\mathbf{x}$ принадлежность к классу опередляется: $s_i=\text{sign} x_i$



__Спектральный метод поиска разреза__

* Используем метод Лагранжа для поиска условного экстремума:
$$Q(\mathbf{x})=\frac{1}{4}\mathbf{x^T}\mathbf{L}\mathbf{x}-\lambda (\mathbf{x^T}\mathbf{x}-n),\text{, при условии } \mathbf{x^T}\mathbf{e}=0$$


* Ищем точку экстремума: продифференциируем по $\mathbf{x}$ и приравняем 0:
$$\mathbf{L}\mathbf{x}=\lambda \mathbf{x}\text{, при условии }\mathbf{x}\perp\mathbf{e}$$


* Пуcть $\mathbf{x}_i$ решение: собственный вектор, отвечающий с.з. $\lambda_i$ и удовлетворяющий $\mathbf{x^T}\mathbf{x}=n$:
$$Q(\mathbf{x}_i)=\frac{n}{4}\lambda_i \rightarrow \min$$


* Наименьшее собственное значение не подходит:
$\lambda_1=0$, $\mathbf{x}_1=\mathbf{x} \Rightarrow \mathbf{x}\not\perp\mathbf{e}$ 


* Берем второе снизу собственное значение $\lambda_2$ и отвечающий ему собственный вектор преобразуется в индекаторный вектор для отнесения к подмножествам по итогам разреза
    * Т.к. для несвязного графа $\lambda_2=0$ (кратность с.з. 0 $\ge 2$) $\Rightarrow$ наименьший разрез $Q(\mathbf{x}_2)=0$
    * <em class="hn"></em> когда имеется малый разрез $Q(\mathbf{x}_2)$, то значение второго снизу с.з. $\lambda_2$ мало

__Спектральный алгоритм поиска разреза__

<em class="cb">Спектральный алгоритм поиска разреза:</em>
* <em class="cr">S1</em> Построить Матрицу Лапласа $\mathbf{L}=\mathbf{D}-\mathbf{A}$ где $D_{ij}=diag(k_i)$
* <em class="cr">S2</em> Найти собственный вектор $\mathbf{x}_2$, отвечающий втрому снизу с.з. $\lambda_2 \ne 0$
* <em class="cr">S3</em> Проеобразовать собственный вектор в индикаторный $\mathbf{s}=f(\mathbf{x}_2, \theta)$
* <em class="cr">S4</em> Выбрать оптимальное значение параметра $\theta$

Базовый вариант: $f=\text{sign}$, настройка параметра $\theta$ не требуется

Вычислительная сложность поиск собственного вектора:
* точное определение собственных значений требует $O(n^3)$
* приближенное значение может получаться существанно быстрее например, с помощью Lanczos method
    * объем вычислений снижается для разреженных матриц (с большим количеством 0)
    * с увеличением разрыва между рассматриваемым и следующим с.з.: $|\lambda_{k+1} - \lambda_{k}|$

Нормализованные vs. ненормализованные Лапласианы
* Если узлы имеют одинаковые или близкие степени - между двумя подходами нет большой разницы
* Если существует большая неоднородность степеней узлов - выбор Лапласиана существенно влияет на результат
* В общем нормализованный Лапласиан является более многообещающим:
    * Ненормализованный Лапласиан учитывает только плотность связей между группами
        * Не всегда сходится + может давать тривиальное разбиение с 1м узлом
    * Нормализованный Лапласиан учитывает плотность по двум наравлениям: как плотность связей между группами, так и плотность связей внутри группы

Lrw is more reliable
than Lsym because the eigenvectors of Lrw corresponding to the lowest eigenvalues are cluster indicator vectors, i.e., they
have equal non-vanishing entries in correspondence of the vertices of each cluster, and zero elsewhere, if the clusters are
disconnected. The eigenvectors of Lsym, instead, are obtained by (left-) multiplying those of Lrw by the matrix D1/2: in this
way, eigenvector components corresponding to vertices of the same cluster are no longer equal, in general, a complication
that may induce artifacts in the spectral clustering procedure.        



The last issue we want to point out concerns the choice of the Laplacian matrix to use in the applications. If the graph
vertices have the same or similar degrees, there is no substantial difference between the unnormalized and the normalized
Laplacians. If there are big inhomogeneities among the vertex degrees, instead, the choice of the Laplacian considerably
affects the results. In general, normalized Laplacians are more promising because the corresponding spectral clustering
techniques implicitly impose a double optimization on the set of partitions, such that the intracluster edge density is high
and, at the same time, the intercluster edge density is low. 

On the contrary, the unnormalized Laplacian is related to the
intercluster edge density only. 

Moreover, unnormalized spectral clustering does not always converge, and sometimes yields
trivial partitions in which one or more clusters consist of a single vertex. 

Of the normalized Laplacians, Lrw is more reliable
than Lsym because the eigenvectors of Lrw corresponding to the lowest eigenvalues are cluster indicator vectors, i.e., they
have equal non-vanishing entries in correspondence of the vertices of each cluster, and zero elsewhere, if the clusters are
disconnected. The eigenvectors of Lsym, instead, are obtained by (left-) multiplying those of Lrw by the matrix D1/2: in this
way, eigenvector components corresponding to vertices of the same cluster are no longer equal, in general, a complication
that may induce artifacts in the spectral clustering procedure.

## Спектральный подход к оптимизации модулярности

__Спектральная максимизация модулярности__

Мера модулярности $^1$ для разбиения сети на сообщества ($S(n_i)$ - сообщество узла $n_i$):
$$Q=\frac{1}{2 L}\sum_{i,j} \left(A_{ij}-\frac{k_i k_j}{2 L}\right) \delta(S(n_i),S(n_j)) \rightarrow \max$$

* Рассматрим случай двух сообществ: $S^{+}$ и $S^{-}$ с индикаторным вектором:
$$s(i) = \begin{cases} \ \ +1 \text{, если } n_i \in S^{+}  \\ \ \ -1 \text{, если } n_i \in S^{-} \end{cases}$$
* <em class="hn"></em> можно записать: $\delta(S(n_i),S(n_j))=\frac{1}{2}(s_i s_j + 1)$
* <em class="hn"></em> функцию модулярности можно записать: 
$$Q=\frac{1}{4 L}\sum_{i,j} \left(A_{ij}-\frac{k_i k_j}{2 L}\right) (s_i s_j + 1)=\frac{1}{4 L} \sum_{i,j} B_{ij} s_i s_j$$,
где: $B_{ij}=\left(A_{ij}-\frac{k_i k_j}{2 L}\right)$
* <em class="hn"></em> функцию модулярности можно записать в виде квадратичной формы :$Q(\mathbf{s})=\frac{1}{4 L} \mathbf{s}^T\mathbf{B}\mathbf{s}\rightarrow \max$
* данная постановка является задачей целочисленной оптимизации и <em class="cb">является NP-сложной</em>

__Непрерывный вариант задачи максимизации модулярности__

* <em class="cb">Приближенная оптимизация</em> - решение аналогичной задачи оптимизации с ослабленными условиями. <em class="cg">Ослаблим требование на получение целочисленного решения.</em> $s \rightarrow x \in \mathbb{R}$
* при этом введем ограничение: $\Vert x \Vert^2=\mathbf{x}^T\mathbf{x}=n$
* <em class="hn"></em> задача квадратичной оптимизации: $Q(\mathbf{x})=\frac{1}{4 L} \mathbf{x}^T\mathbf{B}\mathbf{x}-\lambda(\mathbf{x}^T\mathbf{x}-n)\rightarrow \max$


* Учитывая решение задачи нахождения собственного вектора: $\mathbf{B}\mathbf{x}_i=\lambda_i x_i$
* <em class="hn"></em> имеем приближенное решение задачи максимизации модулярности: $Q(\mathbf{x})=\frac{n}{4 L} \lambda_i$
    * максимум достигается при $\lambda=\max(\lambda_i)$

<em class="cb">Алгоритм максимизации модулярности:</em> 
* <em class="cr">S1</em> Построить матрицу $\mathbf{B}=\mathbf{A}-\frac{1}{2 L}\mathbf{k}\mathbf{k}^T$
* <em class="cr">S2</em> Найти максимальный собственный вектор $\mathbf{x}_{max}$ для задачи $\mathbf{B}\mathbf{x}_i=\lambda_i x_i$
