```python
def makeCycleGraph(n):
    G = nx.Graph()
    G.add_edges_from([(x, (x + 1) % n) for x in range(n)])
    nx.draw(G, with_labels=True)
    plt.show()

makeCycleGraph(5)
```

```python

def lattice(n,m):
    z = nx.Graph()
    y = 1+np.arange(n*m).reshape(n, m)
    h = []
    for i in y:
        for j in i:
            h.append(j)
        
    print(h)
    
    for i in h:
        if i+m<len(h)+1:
            print(i)
            z.add_edges_from([(i, i+m)])
        if i%m != 0:
            z.add_edges_from([(i, i+1)])

    nx.draw(z,with_labels = True)
    plt.show()
    
lattice(4,6)
```

```python
def lattice_adds(n,m,z_1,z_2):
    z = nx.Graph()
    y = 1+np.arange(n*m).reshape(n, m)
    h = []
    for i in y:
        for j in i:
            h.append(j)
        
    print(h)
    
    for i in h:
        if i+m<len(h)+1:
            z.add_edges_from([(i, i+m)])
        if z_1 == True:
            if i%m == 1:
                print('Проверка один:{},{}'.format(i,i+m-1))
                z.add_edges_from([(i, i+m-1)])
        if i%m != 0:
            z.add_edges_from([(i, i+1)])
        if z_2 == True:
            if i<m+1:
                print('Проверка два:{},{}'.format(i,i+(m*n-m)))
                z.add_edges_from([(i, i+(m*n-m))])
            

    nx.draw(z,with_labels = True)
    plt.show()
    
lattice_adds(3,5,0,1)
```

```python
import numpy as np
def fractal_1(n):
    R = nx.Graph([[1,2],[2,3],[3,1]])
    counter = 0
    if n == 1:
        nx.draw(R)
    else:
        b = []
        while counter+1<n:
            a = []
            for i in R.edges():              
                a.append(list(i))
            for i in a:
                q = sum(i)/2+np.random.randint(10,1000)
                b.append([i[0],q])
                b.append([q,i[1]])
                b.append(i)
                R = nx.Graph()
                R = nx.Graph(b)
            a =[]
            counter+=1
        T = nx.Graph(b)
        nx.draw(T, with_labels=True)
fractal_1(2)
```

```python
def fractal_2(n):
    R = nx.Graph([[1,2],[2,3],[3,4],[4,1]])
    counter = 0
    if n == 1:
        nx.draw(R)
    else:
        b = []
        while counter+1<n:
            a = []
            for i in R.edges():              
                a.append(list(i))
            for i in a:
                q = sum(i)/2+np.random.randint(10,1000)
                s = sum(i)/2-np.random.randint(10,1000)
                b.append([i[0],q])
                b.append([q,i[1]])
                b.append([i[0],s])
                b.append([s,i[1]])
                R = nx.Graph()
                R = nx.Graph(b)
            a =[]
            counter+=1
        T = nx.Graph(b)
        print(T.degree())
        for i in T.nodes():
            T.add_nodes_from([i], degree = T.degree(i))
        print(T.nodes[4])
        nx.draw(T, with_labels=True)
fractal_2(2)
```
