# Основы работы с pandas

```python
import numpy as np
import pandas as pd
```

```python
# 1.Получить файл набора данных, прилагающийся к этой работе dataset_1_Titanic.csv(60 КБ)
# 2. Считать файл в pandas DataFrame

data = pd.read_csv('titanic.csv')
data.head()
```

```python
# 3. Оновные параметры датасета:
data.shape # количество измерений и атрибутов
```

Имеем 891 измерение и 12 атрибутов

```python
data.info() # тип данных каждого поля, количество отсутствующих значений для каждого поля,
# количество и названия константных и уникальных полей
```

```python
# 4. Удалить из набора поля, не несущие информативной нагрузки - содержащие уникальные либо константные значения
data = data.drop_duplicates() # удаление повторов
# data
```

```python
data.nunique()  # количество уникальных значений
```

```python
# 7. Заменить все отсутствующие данные средним значением (в случае числового поля) либо самым распространенным 
#    в категориальных полях.
import math 
def isnan(value): 
    try: 
        return math.isnan(float(value))
    except: 
        return False
    
for i in data:    # заходим на уровень колонок
    if data[i].dtype == object:     # если в колонке строки
        most_risent = data[i].value_counts().index[0]    # находим самое распространенное значение
        for el in range(0, len(data[i])) :   # перебираем элементы колонок
            if isnan(data[i][el]) == True:                  
                data[i][el] = most_risent    # заменяем nan
    if data[i].dtype == 'float64' or data[i].dtype == 'int64':     # если в колонке числа
        mn = data[i].mean()     # считаем среднее по колонке
        for el in range(0, len(data[i])) :   # перебираем элементы колонок
            if isnan(data[i][el]) == True:
                data[i][el] = mn   # заменяем nan
data
```

```python
# 5. Заменить возраст пассажиров на год рождения

# 1912 год - год когда затонул Титаник
# данные в графе Age даны на момент его крушения
for i in range (0,len(data.Age)):
    data.Age[i] = int(data.Age[i]+1912)
    if data.Age[i] == 1912:
        data.Age[i] = 0
data.Age
```

```python
# 6. Переименовать все названия колонок на русском языке
data.rename(columns={'PassengerId': 'ПассажирID', 'Survived': 'Выжил', 'Pclass':'Класс', 'Name':'Имя', 'Sex':'Пол', 'Age':'Возраст', 'SibSp':'Колич_родствен',
       'Parch':'Родствен_на_борту', 'Ticket':'Номер_билета', 'Fare':'Цена_билета', 'Cabin':'Номер_каюты', 'Embarked':'Порт_посадки'}, inplace=True)
data.head()
```

```python
# 8. Сохранить получившийся набор данных в формате CSV
data.to_csv('foo.csv', sep = '\t', encoding = 'utf-16', index=False)
```
