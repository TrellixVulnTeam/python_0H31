Таким образом для числовых параметров используем методы Pearson's и Spearman's.

```python
# Pearson's
# По сути мы уже воспользовались этим методом, когда рисовали тепловую карту для корреляционной матрицы.
# Сейчас более четко определим границы: необходимо отобрать те параметры, коэф. корреляции которых 
# с целевой переменной больше 0.5

# Тепловая карта
heatmap_plot = sns.heatmap(corr , vmin=-1, vmax=1, annot=True, cmap="YlGnBu")
plt.show()
```

```python
def PearsonsFS(df):
    arr = df.columns.drop('price')
    res = []
    for k in range(0, len(arr)):
        if df['price'][k]>0.5 and 'price'!= arr[k]:
                   res.append(arr[k])
    return print('PearsonsFS results: ', res)
PearsonsFS(corr)
```

```python
# Этот отбор так же можно сделать с помощью модуля f_regression
from sklearn.feature_selection import f_regression
X = data.iloc[:, [1,5,6,8,9,10,11]]
y = data.price
F, pval = f_regression(X,y)
print('F: ', F)
print('p_value: ', pval)
```

Здесь для второго и третьего параметров p_value больше 0.5, а занчит не является значимым.

Таким, образом отбираем все параметры кроме второго и третьего - тот же результат, что и полученный вручную.

```python
# Spearman's
from scipy.stats import spearmanr
rho, pval = spearmanr(data.drop(columns=['Unnamed: 0']))
# print('Rho: ', rho)
# print('p_value: ', pval)
print(data.drop(columns=['Unnamed: 0']).columns)
pval[6]
```

С помощью этого метода отбираем ['carat', 'table', 'x', 'y','z', 'size']. От предыдущего отличается только тем, что table в этот раз вошло в список значимых параметров.

В итоге для числовых параметров оставляем: ['carat', 'x', 'y', 'z']. 'depth', 'table' - были удалены в процессе feature selection. 'size' - искуственно введенная переменная. Опыт с ее введением не удался, потому что все три переменных  'x', 'y','z' показали свою значимость. 

```python
data.drop(columns=['Unnamed: 0', 'depth','table','size'], inplace=True)
print(data)
```

```python
# Перейдем к отбору категорированных параметров. У нас их три ['cut', 'color', 'clarity']
# Сначала необходимо перейти к числовым значениям.
from sklearn import feature_extraction
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(['Ideal', 'Premium', 'Very Good', 'Good', 'Fair'])
data.cut = le.transform(data.cut)
le.fit(['D', 'E', 'F', 'G', 'H', 'I', 'J'])
data.color = le.transform(data.color)
le.fit(['IF', 'VVS1' ,'VVS2' ,'VS1' ,'VS2', 'SI1', 'SI2', 'I1'])
data.clarity = le.transform(data.clarity)
print(data)
```

```python
# ANOVA
from sklearn.feature_selection import f_classif
X = data.iloc[:,[1,2,3]]
y = data.iloc[:, 4]
F, pval =f_classif(X, y)
print('ANOVA results: ',pval)
```

Здесь для первого параметра p_value больше 0.5, а занчит не является значимым.


```python
# Kendall's
from scipy.stats import kendalltau
cut = data.iloc[:,1]
color = data.iloc[:,2]
clarity = data.iloc[:,3]
y = data.iloc[:, 4]
tau1, _ = kendalltau(cut, y)
print('cut - Tau: ', tau1)
tau2, _ = kendalltau(color, y)
print('color - Tau: ', tau2)
tau3, _ = kendalltau(clarity, y)
print('clarity - Tau: ', tau3)
# print(data.drop(columns=['carat', 'x', 'y', 'z']).columns)
# pval[3]
```

Данный этап не выявил корреляции между всеми категориальными переменныи и целевой переменной. С следствие этого придется, избавиться от этих переменных.

```python
data.drop(columns=['cut', 'color','clarity'], inplace=True)
print(data)
```

МОДЕЛИРОВАНИЕ

```python
# Разделим выборку на тестовую и тренировочную, проведем процедуру стандартизации данных
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

# Split
X, y = d= data.iloc[:,[0,2,3,4]],  data.iloc[:,1]
# Test set will be the 25% taken randomly
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=12)
    
# Standarize
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

print(pd.DataFrame(X_train))

# Статистические параметры тренировочной и тестовой выборок
print ('Training set mean:{:.2f} and standard deviation:{:.2f}'.format(np.average(X_train),np.std(X_train)))
print ('Testing set mean:{:.2f} and standard deviation:{:.2f}'.format(np.average(X_test),np.std(X_test)))
```

Объем выборок большой, так что статистические почти идентичны

```python
# Функция для кривой обучения
from sklearn.model_selection import learning_curve
def plot_learning_curve (estimator, X, y):
    plt.figure()
#     plt.title(title)
    # if ylim is not None:
    #     plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training set")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Tested set")

    plt.legend(loc="best")
    plt.show()
```

```python
# Построим простую линейную модель 
from sklearn import linear_model 
from sklearn import metrics
clf = linear_model.LinearRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
# Кривая обучения для метода опорных векторов с линейным ядром (SVC)
plot_learning_curve (clf, X_test, y_test)
```

О метриках. Буду использовать следующие метрики:
- R^2- Коэффициент детерминации. Указывает на достоверность соответствия набора прогнозов фактическим значениям.- МАЕ - Средняя абсолютная ошибка. Сумма абсолютных различий между прогнозами и фактическими значениями. Это дает представление о том, насколько неправильными были прогнозы, дает представление о величине ошибки, но не дает представление о направлении (например, сверх или при прогнозировании)
- RMSE - Среднеквадратичное отклонение или среднеквадратичная ошибка. Мера различий между значениями, предсказанными моделью, и наблюдаемыми значениями.

```python
# Метрики
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from math import sqrt
r2_1 = r2_score(y_test, y_pred)
print('R2 (1)', r2_1)
rmse1 = sqrt(mean_squared_error(y_test, y_pred))
print('Test RMSE (1): %.3f' % rmse1)
```

```python
# Метод градиентного спуска
from sklearn.linear_model import SGDRegressor
sgd = SGDRegressor()
sgd.fit(X_train, y_train)
y_pred2 = sgd.predict(X_test)
# Кривая обучения 
plot_learning_curve (sgd, X_test, y_test)
# Метрики
r2_2 = r2_score(y_test, y_pred2)
print('R2 (2)', r2_2)
rmse2 = sqrt(mean_squared_error(y_test, y_pred2))
print('Test RMSE (2): %.3f' % rmse2)
```

Метод опорных векторов использовать не буду, потому что он ну оочень долго обучается

```python
# Метод ближайших соседей
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor()
knr.fit(X_train, y_train)
y_pred3 = knr.predict(X_test)
# Кривая обучения 
plot_learning_curve (knr, X_test, y_test)
# Метрики
r2_3 = r2_score(y_test, y_pred3)
print('R2 (3)', r2_3)
rmse3 = sqrt(mean_squared_error(y_test, y_pred3))
print('Test RMSE (3): %.3f' % rmse3)
```

```python
# Деревья решений
from sklearn import tree
dtr = tree.DecisionTreeRegressor()
dtr.fit(X_train, y_train)
y_pred4 = dtr.predict(X_test)
# Кривая обучения 
plot_learning_curve (dtr, X_test, y_test)
# Метрики
r2_4 = r2_score(y_test, y_pred4)
print('R2 (4)', r2_4)
rmse4 = sqrt(mean_squared_error(y_test, y_pred4))
print('Test RMSE (4): %.3f' % rmse4)
```

```python
# Лассо
from sklearn.linear_model import Lasso
reg = Lasso()
reg.fit(X_train, y_train)
y_pred5 = reg.predict(X_test)
# Кривая обучения 
plot_learning_curve (reg, X_test, y_test)
# Метрики
r2_5 = r2_score(y_test, y_pred5)
print('R2 (5)', r2_5)
rmse5 = sqrt(mean_squared_error(y_test, y_pred5))
print('Test RMSE (5): %.3f' % rmse5)
```
