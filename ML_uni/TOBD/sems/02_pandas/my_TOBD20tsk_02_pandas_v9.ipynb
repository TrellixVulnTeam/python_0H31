2.4 Для тикера AAPL подсчитать, сколько дней прошло между максимальным и минимальным значением цены акции на закрытии рынка.

2.5 Для каждого тикера подсчитать, максимальную цену акции и определить, в какой день недели чаще всего достигается максимум цены акции.

```python
tickers = np.unique(dt2['ticker']) # собираем все тикеры
dt2['weekday'] = list(map(lambda i: i.weekday(), dt2['date'])) # столбец с днем недели (в виде числа)
dt22 = dt2.set_index('ticker') # сортируем по тикеру
result = {}
for m in range(len(tickers)):
    line = dt22.loc[tickers[m]].iloc[:,7] # series с днями недели по тикеру
    line = line.reset_index()
    weekdays_counter = {0:0, 1:0, 2:0, 3:0, 4:0} # словарь-счетчик повторений максимумов
    for i in range (0, len(line['weekday'])-4):
        if line['weekday'][i]==0 and line['weekday'][i+4]==4:
            k = dt2['high'][i:i+4].argmax() # находим день недели с максимумом за неделю
            val = weekdays_counter.get(k)
            weekdays_counter.update({k:val+1}) # обновляем счетчик
    max_day = max(list(weekdays_counter.values())) # находим день недели с максимальным числом повторений максимумов по тикеру
    result[tickers[m]] = list(weekdays_counter.keys())[list(weekdays_counter.values()).index(max_day)]
result
```
