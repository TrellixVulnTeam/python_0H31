1. Загрузить датасет avocado. 
2. Преобразовать столбец Date в формате datetime
3. При помощи метода join создать новый датафрейм, в котором столбцами являются скользящие средние цен за 10 дней, а индексом - даты
4. Изобразить на одной фигуре графики полученных временных рядом

```python
avocado = pd.read_csv('data/avocado_v1.csv')
avocado['Date'] = pd.to_datetime(avocado['Date'])

df = pd.DataFrame()

for region in avocado['region'].unique():
    region_df = avocado.copy()[avocado['region']==region]
    region_df.set_index('Date', inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f'{region}_Price10MA'] = region_df['AveragePrice'].rolling(10).mean()
    
    if df.empty:
        df = region_df[[f'{region}_Price10MA']]
    else:
        df  = df.join(region_df[f'{region}_Price10MA'])
```

__2.12 Разбить датафрейм, полученный в Объединение.10, на 4 файла, в зависимости от принадлежности значения столбца 'UnitPrice' интервалам:__

1) $ UnitPrice \in (q_0, q_1] $

2) $ UnitPrice \in (q_1, q_2] $

3) $ UnitPrice \in (q_2, q_3] $

4) $ UnitPrice \in (q_3, q_4] $
    

```python
qs = ec['UnitPrice'].quantile((0, 0.25, 0.5, 0.75, 1)).values
for name, group in ec.groupby(pd.cut(ec['UnitPrice'], qs)):
    print(str(name))
    group.to_csv(f'data/split_q/{str(name)[1:-1]}.csv', index=False, encoding='utf8')
```

__Создать новый столбец UnitPriceWithMean, значения которого совпадают с UnitPrice, если UnitPrice >= 2 и равны среднему значению цен товаров, больших двух, в этой стране, в противном случае__

```python
def replace(group):
    mask = group < 2
    group[mask] = group[~mask].mean()
    return group
```

```python
ec['UnitPriceWithMean'] = ec.groupby(['Country'])['UnitPrice'].transform(replace)
```
