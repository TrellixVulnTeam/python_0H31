## Семинар 1

```python
import os
import zipfile

import dask.bag as db
import re
```

```python
#Распаковка файлов
def extract_files():
    path_zip = os.listdir('D:\Big Data\К')
    for i in path_zip:
        path_fold = os.listdir("D:\Big Data\К\{}".format(i))
        for j in path_fold:
            try:
                zip_ref = zipfile.ZipFile("D:\\Big Data\К\{0}\{1}".format(i, j), 'r')
                zip_ref.extractall("D:\Big Data\K_unzip")
                zip_ref.close()
            except Exception as e:
                print(i + ": " + j)
```

```python
extract_files()
```

1) Подсчитать, сколько раз встречается каждая из русских букв в текстовых файлах (заглавные и строчные символы не различаются), лежащих в папке: https://yadi.sk/d/4Y2h0pKw_taJxQ . Выполнить задание с использованием Dask.Bag .

```python
def absoluteFilePaths(directory):
    for dirpath,_,filenames in os.walk(directory):
        for f in filenames:
            yield os.path.abspath(os.path.join(dirpath, f))
```

```python
def isrus(x):
    return re.match("[а-яА-Я]", x)
```

```python
files_list = list(absoluteFilePaths("../../data/all_k/"))
```

```python
b = db.read_text(files_list, encoding='cp1251', errors = "ignore")
```

```python
%%time
result = b.map(lambda x: x.lower().strip()).flatten().filter(isrus).frequencies().compute()
```

```python
result
```

2) Подсчитать, сколько раз встречается каждое из русских слов (слов, состоящих только из русских букв и знака "-") в текстовых файлах (заглавные и строчные символы не различаются, слова перенесенные через строки не рассматриваются), лежащих в папке: https://yadi.sk/d/4Y2h0pKw_taJxQ .

```python
def isrus(x):
    return re.match("[а-яА-Я]+", x)
```

```python
b = db.read_text("../../data/all_k/1.txt", encoding='cp1251')
```

```python
result = b.map(lambda x: re.sub('[!@#$.?:;\n"]',"",x.lower().strip()).split()).flatten().filter(isrus).frequencies().compute()
```

## Каждая книга в одном элементе Bag

```python
def absoluteFilePaths(directory):
    for dirpath,_,filenames in os.walk(directory):
        for f in filenames:
            yield os.path.abspath(os.path.join(dirpath, f))
```

```python
def isrus(x):
    return re.match("[а-яА-Я]+", x)
```

```python
files_list = list(absoluteFilePaths("../../data/all_k/"))
```

```python
def concat_files_bag():
    b = db.from_sequence([])
    for i in files_list:
        try:
            with open(i, "r", encoding='cp1251') as g:
                    b1 = db.from_sequence([g.read()])
                    b = db.concat([b, b1])
        except Exception as e:
            print(i)
            continue
            
    return b
```

```python
b = concat_files_bag()
```

```python
#Подсчет букв
result = b.map(lambda x: x.lower().strip()).flatten().filter(isrus).frequencies().compute()
```
