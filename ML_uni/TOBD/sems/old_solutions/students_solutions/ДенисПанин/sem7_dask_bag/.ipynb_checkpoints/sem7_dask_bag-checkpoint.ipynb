```python
import glob
import dask.bag as db
import zipfile
import os
import re
import pickle
```

1) Подсчитать, сколько раз встречается каждая из русских букв в текстовых файлах (заглавные и строчные символы не различаются), лежащих в папке: https://yadi.sk/d/4Y2h0pKw_taJxQ . Выполнить задание с использованием Dask.Bag .

```python
# Windows path
path_to_zips = r'..\data\К\*\*.zip'
extract_dir = r'..\data\K'
txt_dir = r'..\data\K\*.TXT'
```

```python
zip_list = glob.glob(path_to_zips, recursive=True)
```

```python
if not os.path.exists(extract_dir):
    os.makedirs(extract_dir)

for zip_file in zip_list:
    with zipfile.ZipFile(zip_file,"r") as zip_ref:
        try:
            zip_ref.extractall(extract_dir)
        except Exception:
            pass
```

```python
txt_list = glob.glob(txt_dir, recursive=True)
txt_list
```

dask

```python
alphabet = set("абвгдеёжзийклмнопрстуфхцчшщъыьэюя")
def is_char(char):
    return char in alphabet
```

```python
%%time
b = db.from_sequence(txt_list).map(db.read_text, encoding='cp1251', errors='ignore')
lower_files = b.map(lambda file: file.str.lower())
computed_char_bags = lower_files.map(lambda file: file.flatten())
filtered_chars = computed_char_bags.map(lambda b: b.filter(is_char)).compute()
one_bag = db.concat(filtered_chars)
b1_result = one_bag.frequencies().compute()
```

```python
%%time
b = db.from_sequence(txt_list).map(db.read_text, encoding='cp1251', errors='ignore').compute()
b = db.concat(b)
lower_files = b.str.lower()
computed_char_bags = lower_files.flatten()
one_bag = computed_char_bags.filter(is_char)
b11_result = one_bag.frequencies().compute()
```

```python
b11_result
```

```python
with open("b1_result.pickle", 'wb') as f:
    pickle.dump(b1_result, f)
```

2) Подсчитать, сколько раз встречается каждое из русских слов (слов, состоящих только из русских букв и знака "-") в текстовых файлах (заглавные и строчные символы не различаются, слова перенесенные через строки не рассматриваются), лежащих в папке: https://yadi.sk/d/4Y2h0pKw_taJxQ .

```python
b2 = db.from_sequence(txt_list).map(db.read_text, encoding='cp1251', errors='ignore')
```

```python
def clear_words(word):
    w = word.strip()
    return re.sub('[!?.,:]', '', w)
```

```python
lower_files = b.map(lambda file: file.str.lower())
bags_of_words_and_symbols = lower_files.map(lambda x: x.str.split(' ')).map(lambda x: x.flatten())
bags_of_words = bags_of_words_and_symbols.map(lambda a: a.filter(lambda x: re.match(r"\w+", x)))
bags_of_clean_words = bags_of_words.map(lambda b: b.map(clear_words))
```

```python
b2_result = db.concat(bags_of_clean_words.compute()).frequencies().compute()
```

```python
with open("b2_result.pickle", 'wb') as f:
    pickle.dump(b2_result, f)
```
