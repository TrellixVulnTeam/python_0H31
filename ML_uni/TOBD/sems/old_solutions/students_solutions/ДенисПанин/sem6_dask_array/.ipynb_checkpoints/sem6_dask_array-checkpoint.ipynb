3) Сгенирировать нормально распределенный одномерный массив из 1000 элементов (вектор $a$). Найти ближайший к нему (в смысле L2) вектор-столбец из исходного массива (вектор $v_n$).

```python
a = da.random.normal(0, 1, size=(1000,1), chunks=(1000,1))
```

```python
a_l2 = da.sum(da.square(a))
```

```python
x_l2 = da.sum(da.square(x), axis=0)
```

```python
distances = da.absolute(x_l2-a_l2)
```

```python
vn_ind = da.argmin(distances).compute()
```

4) Определить сколько векторов-столбцов из исходного массива находятся на расстоянии, не превышающем $3|a-v_n|_{L2}$

```python
vn = x[:, vn_ind]
vn_l2 = da.sum(vn**2)
```

```python
max_dist = 3*da.absolute(a_l2 - vn_l2)
```

```python
answer = da.sum(distances <= max_dist).compute()
```
