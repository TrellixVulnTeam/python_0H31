Расчитать количество букв в текстовом файлу

```python
import dask.bag as db
import os 
import zipfile
```

```python
import collections
def read_file(dir_file):
    c = collections.Counter()
    list_rus_symb = 'ёйцукенгшщзхъфывапролджэюбьтимсчя'
    text = db.read_text(dir_file, encoding='cp1251').compute()
    
    for sent in text:
        for i in sent:
            if i.lower() in list_rus_symb:
                c[i.lower()] += 1
    return c
```
