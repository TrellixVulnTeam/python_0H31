### Удаление из папки всех файлов с "плохой" кодировкой

```python
import chardet
data=r'C:\\Users\\Big data\\New\\all_k'
not_correct = []

bad_files = "bad_file"

if not os.path.exists(os.path.dirname(bad_files)):
    os.makedirs(bad_files)

files = os.listdir(data)

for file in files:
    file_full = os.path.join(data, file)
    with open(file_full, 'rb') as f:
        rawdata = b''.join([f.readline() for _ in range(20)])
    encoding = chardet.detect(rawdata)['encoding']
#    print(f"{file} -> {encoding}")
    if encoding != 'windows-1251':
        not_correct.append((file, encoding))
        os.rename(file_full, os.path.join(os.getcwd(), bad_files, file))
```

### 2) Посчитать слова в тексте

```python
import string
```

```python
punc = string.punctuation.replace('-', '')
table = str.maketrans('', '', punc)
```

```python
files = [os.path.join('C:\\Users\\Big data\\New\\all_k', i) for i in os.listdir('C:\\Users\\Big data\\New\\all_k')]
```

```python
def count_words(file_name):
    with open(file_name, encoding='ANSI') as f:
        count = Counter(chain.from_iterable(re.findall('[а-я-]+', line.lower()) for line in f))
    return count
```

```python
%%time
db.from_sequence(files).map(count_words).fold(op.add).compute()
```

```python
%%time
db.from_sequence(files).map(count_words).flatten().frequencies().compute()
```

```python
def count_words2(file_name):
    with open(file_name, encoding='ANSI') as f:
        count = Counter(chain.from_iterable(line.translate(table).lower().split() for line in f))
    return count
```

```python
%%time
db.from_sequence(files).map(count_words2).fold(op.add).compute()
```

### Для одного из файлов с отзывами из https://yadi.sk/d/OUpjPvLX0i1Ipw подсчитать частоту (!) слов для отзывов с оценкой "5" и для отзывов с оценкой "2" и найти 50 распространенных слов (встречающихся в каждой из групп более 10 раз), частота которых максимально увеличивается и уменьшается при переходе от плохих отзывов к хорошим.

```python
%%time
scores = dict(js.filter(lambda x: x['overall'] == 5).\
                 map(lambda x: re.findall(r'[a-z]+', x['reviewText'].lower())).flatten().frequencies().compute())
```

```python
for i in list(scores):
    if scores[i] < 10: scores.pop(i)
```

```python
sum_pos = sum(scores.values())
len(scores)
```

```python
for i in list(scores.keys()):
    scores[i] /= sum_pos
```

```python
neg_scores = dict(js.filter(lambda x: x['overall'] == 2).\
                 map(lambda x: re.findall(r'[a-z]+', x['reviewText'].lower())).flatten().frequencies().compute())
```

```python
for i in list(neg_scores):
    if neg_scores[i] < 10: neg_scores.pop(i)
len(neg_scores)
```

```python
neg_pos_sum = sum(neg_scores.values())
for i in list(neg_scores.keys()):
    neg_scores[i] /= neg_pos_sum
```

```python
common_keys = set(neg_scores.keys()).intersection(set(scores.keys()))
common_dict = dict()
for i in common_keys:
    common_dict[i] = scores[i] / neg_scores[i]
```

```python
b = sorted(list(common_dict.items()), key=lambda x: x[1])
b[-50:]
```
