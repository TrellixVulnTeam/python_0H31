```python
dt.rename(columns={'Ставка по вкладам в рублях до\xa01\xa0года': 'Ставка по вкладам в рублях до 1 года'}, inplace=True)
dt.columns.tolist()
```

```python
# Разделяем данные на тренирововчную и тестовую
dt1 = pd.DataFrame() # dataframe с известными данными, на которых будем исследовать зависимость
dt2 = pd.DataFrame() # dataframe с данными для тестовой выборки
dt1 = dt.iloc[:dt.shape[0]-12]
dt2 = dt.iloc[dt.shape[0]-12:]
dt1
```

```python
# Визуализация прироста объема вкладов
plt.subplots(figsize = (16,6))
plt.plot(dt1['Прирост вкладов физических лиц в рублях (млн руб)'])
plt.show
```

```python
dt_corr = pd.DataFrame()
dt_corr['Прирост вкладов физических лиц'] = dt1['Прирост вкладов физических лиц в рублях (млн руб)'].iloc[1:]
features = ['Доходность ОФЗ по сроку до 1 года', 'Ключевая ставка',
       'Ставка по вкладам в долларах до 1 года','Ставка по вкладам в рублях до 1 года',
        'Нефть марки Юралс, долл./барр',
       'Индекс потребительских цен, ед.', 'М0, руб', 'М2, руб',
       'Курс доллара к рублю, руб', 'Номинальная средняя з/п, руб']
for feature in features:
    dt_corr[feature] = 1
    dt_corr = dt_corr.astype({feature: float})
    for i in range(1,len(dt_corr[feature])+1):
        dt_corr[feature][i] = dt1[feature][i] - dt1[feature][i-1]
    dt_corr[feature]
dt_corr
```

```python
plt.subplots(figsize = (16,10))
sns.heatmap(dt_corr.corr(),vmin=-1, vmax=1, annot=True, cmap="YlGnBu").set_title("Корреляция")
plt.show()
```

```python
items = dt_corr.columns.tolist()
for i in items:
    _,ax = plt.subplots(figsize = (16,6))
    plt.plot(dt_corr[i])
    ax.set_title(i)
    plt.show
```

```python
y = dt_corr['Прирост вкладов физических лиц']
X = dt_corr.drop('Прирост вкладов физических лиц', axis=1)
cols = dt_corr.columns.tolist()
features = cols[1:]

for i in range (0, len(features)):
    plt.figure()
    _,ax = plt.subplots()
    x = dt_corr[features[i]]
    plt.scatter(x, y)
    ax.set_title(features[i])


plt.show()
```
