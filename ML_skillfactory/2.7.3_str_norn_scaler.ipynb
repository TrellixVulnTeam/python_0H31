```python
import pandas as pd
import numpy as np
import sklearn
```

```python
vis_data = pd.read_csv("train.csv", encoding = 'ISO-8859-1', low_memory = False)
vis_data
```

```python
X = vis_data['balance_due'].values.reshape(-1, 1)
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
X_scaled.min().round(5)
```

```python
new_values = []
for i in range(len(X)):
    if X[i]>0:
        new_values.append(np.sqrt(X[i]))
new_values
```

```python
medi = np.median(new_values)
ave = np.mean(new_values)
abs(medi-ave)
```
